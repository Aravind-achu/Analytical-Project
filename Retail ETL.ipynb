```python
da=pd.read_csv('D:\\Analytical project\\orders.csv',na_values=['Not Available','unknown'])
da['Ship Mode'].unique()
```

```python
da.rename(columns={'City':'city','Order Id':'order_id'})
```

```python
da.columns=da.columns.str.lower()
da.columns=da.columns.str.replace(' ', '_')
da.head(3)
```

```python
da['discount']=da['list_price']*da['discount_percent']/100
da.head(6)
```

```python
da['sale_price']=da['list_price']*da['discount']
da.head(6)
```

```python
da['profit']=da['sale_price']-da['cost_price']
da.head(6)
```

```python
da['order_date']=pd.to_datetime(da['order_date'],format='%Y-%m-%d')
da.head(3)
```

```python
da.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
